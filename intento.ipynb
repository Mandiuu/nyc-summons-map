```python
import pandas as pd

# Define the path to the CSV file
csv_file_path = 'vaccination_data.csv'

# Read the CSV file with a semicolon delimiter
vaccination_data = pd.read_csv(csv_file_path, delimiter=';')

# Display the first few rows and the columns of the dataset
print(vaccination_data.head())
print(vaccination_data.columns)
```

```python

# Clean the column names by stripping whitespace and replacing spaces with underscores
vaccination_data.columns = vaccination_data.columns.str.strip().str.replace(' ', '_')

# Convert precinct column to string if it's not already
vaccination_data['precinct'] = vaccination_data['precinct'].astype(str).str.strip()

# Fill any missing values with 0 for numerical columns
for col in vaccination_data.columns:
    if vaccination_data[col].dtype in ['int64', 'float64']:
        vaccination_data[col].fillna(0, inplace=True)

# Display the cleaned dataset
print(vaccination_data.head())

# Save the cleaned data back to CSV (if needed)
cleaned_csv_file_path = 'cleaned_vaccination_data.csv'
vaccination_data.to_csv(cleaned_csv_file_path, index=False)
```
